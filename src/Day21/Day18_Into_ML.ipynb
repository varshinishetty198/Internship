The House Price India dataset contains information about residential properties in India, where each row represents one house and each column describes its features such as price, bedrooms, bathrooms, area, and location

```python
plt.figure()
sns.histplot(df['Price'], kde=True)
plt.title("Distribution of House Prices")
plt.show()
```

```python
df.columns
```

```python
plt.figure()
sns.scatterplot(x='number of bedrooms', y='Price', data=df)
plt.title("Bedrooms vs Price")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
